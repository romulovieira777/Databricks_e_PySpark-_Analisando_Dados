```
%sql
SELECT
  pais,
  SUM(preco) AS total_vendido
FROM
  vinho
WHERE
  preco > 0
GROUP BY
  pais
ORDER BY
  total_vendido DESC
LIMIT
  10
```

```
%sql
SELECT
  pais,
  variante,
  SUM(preco) AS total_vendido
FROM
  vinho
WHERE
  preco > 0
GROUP BY
  pais,
  variante
ORDER BY
  total_vendido DESC
LIMIT
  10
```
